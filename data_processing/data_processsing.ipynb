```python
import scipy.io
import os 
import numpy as np

# # Load MATLAB file
# mat_data = scipy.io.loadmat('mpii_human_pose_v1_u12_1.mat')

# print(mat_data)


matlab_mpii = scipy.io.loadmat('mpii_human_pose_v1_u12_1.mat', struct_as_record=False)['RELEASE'][0, 0]
num_images = annotation_mpii = matlab_mpii.__dict__['annolist'][0].shape[0]
# Access variables in the MATLAB file
```

```python
from tqdm import tqdm
import copy
```

```python
num_images 
```

```python
mpii_idx_to_jnt = {0: 'rankl', 1: 'rknee', 2: 'rhip', 5: 'lankl', 4: 'lknee', 3: 'lhip',
                   6: 'pelvis', 7: 'thorax', 8: 'upper_neck', 11: 'relb', 10: 'rwri', 9: 'head',
                   12: 'rsho', 13: 'lsho', 14: 'lelb', 15: 'lwri'}
mpii_template = dict([(mpii_idx_to_jnt[i], []) for i in range(16)])
```

```python
img_dict = {'mpii': {'img': [], 'img_name': [], 'img_pred': [], 'img_gt': []}}
```

```python
for img_idx in tqdm(range(0,num_images )):
    annotation_mpii = matlab_mpii.__dict__['annolist'][0, img_idx]
    train_test_mpii = matlab_mpii.__dict__['img_train'][0, img_idx].flatten()[0]
    person_id = matlab_mpii.__dict__['single_person'][img_idx][0].flatten()

    # Load the individual image. Throw an exception if image corresponding to filename not available.
    img_name = annotation_mpii.__dict__['image'][0, 0].__dict__['name'][0]
    gt_per_image = copy.deepcopy(mpii_template)
    annotated_person_flag = False
    if len(person_id) == 1:
        for person in (person_id - 1):
            try:
                annopoints_img_mpii = annotation_mpii.__dict__['annorect'][0, person].__dict__['annopoints'][0, 0]
                num_joints = annopoints_img_mpii.__dict__['point'][0].shape[0]

                # Iterate over present joints
                for i in range(num_joints):
                    x = annopoints_img_mpii.__dict__['point'][0, i].__dict__['x'].flatten()[0]
                    y = annopoints_img_mpii.__dict__['point'][0, i].__dict__['y'].flatten()[0]
                    id_ = annopoints_img_mpii.__dict__['point'][0, i].__dict__['id'][0][0]
                    vis = annopoints_img_mpii.__dict__['point'][0, i].__dict__['is_visible'].flatten()

                    # No entry corresponding to visible
                    if vis.size == 0:
                        vis = 1
                    else:
                        vis = vis.item()

                    gt_per_joint = np.array([x, y, vis]).astype(np.float16)
                    gt_per_image[mpii_idx_to_jnt[id_]].append(gt_per_joint)
    #                 print(gt_per_joint)
                annotated_person_flag = True
            except KeyError:
                print("# Person 'x' could not have annotated joints, hence move to person 'y'")
                continue

        if not annotated_person_flag:
            continue

        # Update the template copy with image, name and ground truth
    #     img_dict['mpii']['img'].append(image)
        img_dict['mpii']['img_name'].append(img_name)
        img_dict['mpii']['img_gt'].append(gt_per_image)
```

```python
def generate_text_single(joint_points, image_name=None):
    temp_text = ""
    for key, value in joint_points.items():
        temp_text+= key
        temp_text+=":"
        temp_text+= str(round(value[0][0]))
        temp_text+= ','
        temp_text+= str(round(value[0][1]))
        temp_text+= ' '
    return temp_text
```

```python
def generate_text_single_normalized(joint_points, image_size):
    temp_text = ""
    for key, value in joint_points.items():
        temp_text+= key
        temp_text+=":"
        temp_text+= str(round(value[0][0]/image_size[0],3))
        temp_text+= ','
        temp_text+= str(round(value[0][1]/image_size[1],3))
        temp_text+= ' '
    return temp_text
```

```python
def generate_simple_text_single(joint_points, image_size=None):
    temp_text = ""
    target_key = ['head', 'rknee', 'lknee', 'relb', 'lelb']
    for key, value in joint_points.items():
        if key in target_key:
            temp_text+= key
            temp_text+=":"
            temp_text+= str(round(value[0][0]))
            temp_text+= ','
            temp_text+= str(round(value[0][1]))
            temp_text+= ' '
    return temp_text
```

```python
def generate_simple_text_single_normalized(joint_points, image_size):
    temp_text = ""
    target_key = ['head', 'rknee', 'lknee', 'relb', 'lelb']
    for key, value in joint_points.items():
        if key in target_key:
            temp_text+= key
            temp_text+=":"
            temp_text+= str(round(value[0][0]/image_size[0],3))
            temp_text+= ','
            temp_text+= str(round(value[0][1]/image_size[1],3))
            temp_text+= ' '
    return temp_text
```

```python
def generate_text(joint_dict, image_size):
    for i in range(len(joint_dict["joint_points"])):
        temp_text = generate_text_single(joint_dict["joint_points"][i],image_size[i])
        simple_temp_text = generate_simple_text_single(joint_dict["joint_points"][i], image_size[i])
        temp_text_normal = generate_text_single(joint_dict["joint_points"][i],image_size[i])
        simple_temp_text_normal = generate_simple_text_single(joint_dict["joint_points"][i], image_size[i])
        
        joint_dict['text'].append(temp_text)
        joint_dict['simple_text'].append(simple_temp_text)
        joint_dict['normalized_text'].append(temp_text_normal )
        joint_dict['normalized_simple_text'].append(simple_temp_text_normal)
```

```python
def generate_visibility_single(joint_points):
    temp_text = ""
    for key, value in joint_points.items():
        temp_text+= key
        temp_text+=":"
        temp_text+= str(int(value[0][2]))
#         temp_text+=","
        temp_text+= ' '
    return temp_text
```

```python
def generate_simple_visibility_single(joint_points):
    temp_text = ""
    target_key = ['head', 'rknee', 'lknee', 'relb', 'lelb']
    for key, value in joint_points.items():
        if key in target_key:
            temp_text+= key
            temp_text+=":"
            temp_text+= str(int(value[0][2]))
#             temp_text+=","
            temp_text+= ' '
    return temp_text
```

```python
def generate_visibility_text(joint_dict):
    for i in range(len(joint_dict["joint_points"])):
        temp_text = generate_visibility_single(joint_dict["joint_points"][i])
        simple_temp_text = generate_simple_visibility_single(joint_dict["joint_points"][i])
        joint_dict['visibility_text'].append(temp_text)
        joint_dict['simple_visibility_text'].append(simple_temp_text)
```

```python
def check_all_joint_label_visible(joint_dict):
    mpii_idx_to_jnt = {0: 'rankl', 1: 'rknee', 2: 'rhip', 5: 'lankl', 4: 'lknee', 3: 'lhip',
                   6: 'pelvis', 7: 'thorax', 8: 'upper_neck', 11: 'relb', 10: 'rwri', 9: 'head',
                   12: 'rsho', 13: 'lsho', 14: 'lelb', 15: 'lwri'}
    flag = True
    for value in mpii_idx_to_jnt.values():
        if joint_dict[value] == []:
            flag = False
            break
    return flag
```

```python
def get_processed_data(img_dict):
    result ={'img_name': [], 'joint_points': [], 'text':[], 
                                         'simple_text':[],
                                         'visibility_text':[],
                                        'simple_visibility_text':[],
                                        'normalized_text':[],
                                        'normalized_simple_text':[],
            }
    all_joint_label_visible_index = []

    for i in range(len(img_dict['mpii']['img_gt'])):
        if check_all_joint_label_visible(img_dict['mpii']['img_gt'][i]):
            all_joint_label_visible_index.append(i)

    result['img_name'] = [img_dict['mpii']['img_name'][i] for i in all_joint_label_visible_index]
    result['joint_points'] = [img_dict['mpii']['img_gt'][i] for i in all_joint_label_visible_index]
    image_sizes = []
    print('start size')
    for i in range(len(all_joint_label_visible_index)):
        name = result['img_name'][i]
        temp_image = Image.open(f"./MPII-Human-Pose-Visualization/data/mpii/images/{name}")
        image_sizes.append(copy.copy(temp_image.size))
    print("finish size")
    generate_text(result, image_sizes)
    generate_visibility_text(result)
    #     print(all_joint_label_visible_index)
#     print(all_joint_label_part_visible )
    return result
    
```

```python
result = get_processed_data(img_dict)
```
